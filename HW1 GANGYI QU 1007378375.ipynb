# Step 1: We assume the $\theta_A$ = 0.3 | $\theta_B$ =0.5 | $\theta_C$ = 0.8 | and sample each game 10 times and see the probability

```python
import numpy as np
from scipy import stats

# since we unknown everything we will keep play each game ten times, and see the probability we get.
stats.binom(n=1,p= 0.3).rvs(size=1)
a = 0
for i in range(10):
    if stats.binom(n=1,p= 0.5).rvs(size=1) == 1:
        a = a+1
        
probA = a/10
print(probA)


stats.binom(n=1,p= 0.5).rvs(size=1)
b = 0
for i in range(10):
    if stats.binom(n=1,p= 0.5).rvs(size=1) == 1:

        b = b+1
        
probB = b/10
print(probB)


stats.binom(n=1,p= 0.8).rvs(size=1)
c = 0
for i in range(10):
    if stats.binom(n=1,p= 0.8).rvs(size=1) == 1:

        c = c+1
        
probC = c/10
print(probC)

```

# Step 2: Start with uniform prior and analysis for one roundÂ (Initial Prior)

```python
alpha,beta,n,x=1,1,10,3 # from we select A for 10 round,we win 3 rounds
parameter_domain = np.linspace(0,1,1001)
likelihood = stats.binom.pmf(n=10, k=3, p= parameter_domain )
prior = stats.beta(a=1,b=1).pdf(parameter_domain)
c = 11
joint = c*likelihood*prior
posterior = stats.beta(a=x+alpha,b=n-x+beta).pdf(parameter_domain)
plt.figure(figsize=(12,4)); plt.plot(parameter_domain, likelihood, label="likelihood"); plt.plot(parameter_domain, prior, label="prior"); plt.plot(parameter_domain, posterior, label="posterior"); plt.plot(parameter_domain, joint, linestyle="--", label="joint (times scalar "+str(c)+")"); plt.legend(); plt.title("A from Beta(1,1) to Beta(13,8)" ); 
```
